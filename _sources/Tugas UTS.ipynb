# **KLASIFIKASI NAIVE BAYES**

# **PRE-PROCESSING NAIVE BAYES**
Pre-processing atau normalisasi data adalah proses mengubah data dari bentuk awalnya menjadi format yang dapat diolah oleh algoritma machine learning. Tahap pertama dalam normalisasi data adalah mengubah nilai-nilai kategorikal menjadi nilai numerik dengan menggunakan library Scikit-learn dan melakukan import class LabelEncoder. Setelah itu, dilakukan transformasi label encoding pada setiap kolom dataset dengan memanggil method fit_transform dari objek LabelEncoder. Transformasi label encoding mengubah nilai-nilai kategorikal menjadi nilai numerik. Selanjutnya, tahap preprocessing selanjutnya adalah mengubah nilai numerik dalam dataset menjadi nilai yang berkisar antara 0 dan 1 dengan menggunakan class MinMaxScaler dari library Scikit-learn. Dataset yang sudah dimuat diubah menggunakan transformasi Min-Max Scaling dengan memanggil method fit_transform dari objek MinMaxScaler pada dataset. Transformasi Min-Max Scaling adalah sebuah teknik preprocessing data yang sering digunakan untuk menormalkan nilai-nilai numerik dari suatu dataset agar memiliki rentang nilai yang sama, biasanya rentang 0 - 1.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset = dataset.apply(le.fit_transform)
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
dataset = sc.fit_transform(dataset)
dataset
```

# **MEMBAGI DATA MENJADI DATA LATIH DATA UJI (SPLIT DATA)**

pada tahap ini data dibagi menjadi data latih dan data uji yang mana Fungsi ini digunakan untuk membagi dataset menjadi data latih dan data uji secara acak  sehingga data latih dan data uji dapat merepresentasikan dataset secara keseluruhan.
pada klasifikasi saat ini saya menggunakan split data 70 30 dimana 70% adalah data uji dan 30% data latih 70%

```python
from sklearn.model_selection import train_test_split
X = dataset[: , :-1]
y = dataset[: , -1]
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3 , random_state = 42)
```

# **DATA LATIH**
tahap ini merupakan tahapan latih atau training untuk prediksi pada data uji.
import fungsi GaussianNB dari library Scikit-learn (sklearn.naive_bayes) dan membuat objek GNB dari kelas GaussianNB yang Kemudian, fungsi fit() dipanggil pada objek GNB yang telah dibuat dengan parameter X_train dan y_train, yaitu data latih (variabel independen dan dependen) yang digunakan untuk melatih model.

```python
from sklearn.naive_bayes import GaussianNB
GNB = GaussianNB()
GNB.fit(X_train,y_train)
```

# **PREDIKSI**
tahap ini adalah untuk prediksi data dengan menggunakan fungsi predict()

```python
y_pred = GNB.predict(X_test)
```

# **SKOR AKURASI NAIVE BAYES**
mengecek akurasi nya bisa menggunakan fungsi accuracy_score()

```python
from sklearn.metrics import accuracy_score
acnb = accuracy_score(y_test,y_pred)
```

```python
print('SKOR AKURASI NAIVE BAYES = ', acnb)
```
