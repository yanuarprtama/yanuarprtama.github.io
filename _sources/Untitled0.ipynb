Berikut adalah langkah-langkah yang dapat dilakukan untuk menganalisis data ini dengan menggunakan kedua algoritma tersebut:

1. Import library yang diperlukan

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

2. Muat data dari sumber data dan ubah menjadi dataframe menggunakan pandas

```python
url = "https://raw.githubusercontent.com/datasets/breast-cancer/master/data/breast-cancer.csv"
df = pd.read_csv(url)
df.head(5)
```

3. Mengekstrak fitur dan label dari data

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

4. Bagi data menjadi data pelatihan dan data uji

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
y_pred_nb = nb.predict(X_test)
```

```python
acc_nb = accuracy_score(y_test, y_pred_nb)
report_nb = classification_report(y_test, y_pred_nb)
cm_nb = confusion_matrix(y_test, y_pred_nb)
print("Accuracy Naive Bayes:", acc_nb)
print("Classification Report Naive Bayes:\n", report_nb)
print("Confusion Matrix Naive Bayes:\n", cm_nb)
```
